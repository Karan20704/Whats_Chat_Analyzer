```python
import re
import pandas as pd
```

```python
f = open('WhatsApp Chat with IT_elite_24hour.txt','r',encoding='utf-8')
```

```python
data = f.read()
```

```python
print(data)  #data in string format
```

```python
print(type(data))
```

```python
# converting string format into pandas dataframe
# 1 coloumn ---> date
# 2 coloumn ---> message
```

```python
pattern = '\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}\s-\s'
```

```python
messages = re.split(pattern,data)[1:] # at staring empty string come to remove it [1:]
print(messages)
```

```python
print(type(messages))
```

```python
dates = re.findall(pattern,data)
dates
```

```python
print(type(dates))
```

```python
df = pd.DataFrame({'user_message' : messages , 'message_date' :dates})
df['message_date'] = pd.to_datetime(df['message_date'],format='%d/%m/%Y, %H:%M - ')
df.rename(columns={'message_date' : 'date'},inplace=True)
df.head()
```

```python
# seprate user and message
users = []
messages = []
for message in df['user_message']:
    # Improved regex to handle user names with spaces and special characters
    entry = re.split(r'([\w\s\(\)]+?):\s', message, maxsplit=1)
    if len(entry) > 1:
        users.append(entry[1].strip())
        messages.append(entry[2].strip())
    else:
        users.append('group_notification')
        messages.append(entry[0].strip())

df['user'] = users
df['message'] = messages
df.drop(columns=['user_message'], inplace=True)
```

```python
period = []
for hour in df[['day_name', 'hour']]['hour']:
    if hour == 23:
        period.append(str(hour) + "-" + str('00'))
    elif hour == 0:
        period.append(str('00') + "-" + str(hour + 1))
    else:
        period.append(str(hour) + "-" + str(hour + 1))
```

```python
df['period'] = period
```

```python
df
```

```python
import seaborn as sns
plt.figure(figsize=(20, 6))
sns.heatmap(df.pivot_table(index='day_name', columns='period', values='message', aggfunc='count'))
plt.yticks(rotation='horizontal')
plt.show()
```
